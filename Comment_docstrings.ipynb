### Docstring cho function

Nội dung docstring cho hàm (function) nằm ngay sau khi khai báo tên hàm, và trước mọi dòng code khác:

```python
def my_function(args):
    
    '''
    Docstring cho hàm đặt ở đây
    '''
```

Nội dung của function docstring cần bao gồm những thông tin sau:
    
1) Mô tả ngắn gọn công dụng của hàm: nó dùng để làm gì ?
    
2) Nếu hàm nhận input, cần giải thích về bản chất của input (dataframe, list, dict hay array ?, float hay int hay strings)

3) Giải thích ý nghĩa của tấtcả arguments

4) Nếu hàm xuất kết quả, cần giải thích bản chất output là gì ? Có bao nhiêu output ?

5) Nếu hàm có side effects (thí dụ in bảng biểu, vẽ hình, save dữ liệu, hình ảnh, cần chú thích

6) Một vài thí dụ về syntax

### Docstring cho class

Với class, docstring cần được sử dụng ở cả 2 cấp: dành cho class, và dành cho mỗi method trong class (bao gồm instance method, class method và static method, xem như những functions)

#### Docstring cho class cần cung cấp các thông tin như:

1) Tóm tắt ý nghĩa, công dụng của class : để làm gì, đại diện cho cái gì ?

2) Tóm tắt những attributes của class

3) Tên và công dụng tóm tắt các methods của class

#### Docstring cho methods

Docstring cho method có cấu trúc và nội dung như docstring cho function

```python
class Xet_nghiem:
    
    '''
    Class đại diện cho kết quả xét nghiệm của 1 bệnh nhân
    
    Attributes
    ----------
    patient(str): tên bệnh nhân
    result(float): giá trị xét nghiệm
    normal(float): giá trị tham chiếu
    
    
    Methods
    -------
    print_result: instance method, In kết quả
    blood_test: class method, tạo 1 xét nghiệm máu
    check: static method, kiểm tra thủ công
    
    '''
    
    def __init__(self, patient, result, normal):
        
        '''
        Attributes
        ----------
    patient(str): tên bệnh nhân
    result(float): giá trị xét nghiệm
    normal(float): giá trị tham chiếu
        '''
        
        self.patient = patient
        self.result = result
        self.normal = normal
            
    # Instance method
    
    '''
    In kết quả xét nghiệm và chẩn đoán
    '''
    
    def print_result(self):
        if self.result <= self.normal:
            diag = 'Bình thường'
        else:
            diag = 'Bất thường'
        
        return f'Kết quả {self.exp_type} của bệnh nhân {self.patient}: {result}, kết luận: {diag}'

    @classmethod
    def blood_test(cls):
        
        '''
        Tạo ra 1 class xét nghiệm máu
        
        '''
        
        print('Đây là xét nghiệm máu')
        return cls

    @staticmethod
    def check(a,b):
        
        '''
        Kiểm tra thủ công
        
        a(float): giá trị cần so sánh
        b(float): giá trị tham chiếu
        '''
        if a <= b:
            print('Bình thường')
        else:
            print('Bất thường')
```
